## LLM evaluator

### The prompt

```python
def evaluate_answers(question, enhanced_answer, baseline_answer):
    prompt = f"""
    You are tasked with evaluating the answers provided by two methods for the following question:

    **Question:** {question}

    **Enhanced RAG Answer:** {enhanced_answer}

    **Baseline RAG Answer:** {baseline_answer}

    For each answer, compare the following criteria and specify the winner by a single digit 1(Enhanced RAG) or 2(Baseline RAG):

    1. Comprehensiveness: How much detail does the answer provide to cover all aspects and details of the question?
    2. Diversity: How varied and rich is the answer in providing different perspectives and insights on the question?
    3. Empowerment: How well does the answer help the reader understand and make informed judgments about the topic?
    4. Directness: How specifically and clearly does the answer address the question?

    Please provide your ratings in the following format:

    **Ratings:**
    - Comprehensiveness: [insert winner]
    - Diversity: [insert winner]
    - Empowerment: [insert winner]
    - Directness: [insert winner]

    **Comments:**
    [Your brief comments or justifications for your ratings here.]

    Example Response:
    **Ratings:**
    - Comprehensiveness: 1
    - Diversity: 1
    - Empowerment: 1
    - Directness: 2

    **Comments:**
    The Enhanced RAG Answer provided covered most aspects of the question with good examples, while the Baseline RAG Answer was less detailed and missed some key points.
    """

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content.strip()
```

### Run Evaluation

```python
results = []
for index, question in enumerate(questions):
    proposed_method_answer = proposed_method_result.at[index, 'answer']
    baseline_rag_answer = baseline_rag_result.at[index, 'answer']

    evaluation = evaluate_answers(question, proposed_method_answer, baseline_rag_answer)
    results.append({'Question': question, 'Evaluation': evaluation})
```

### Extract and store the winners

```python
import re
import pandas as pd

def extract_winners(evaluation):
    ratings_pattern = re.compile(r'(\w+):\s*(\d)')
    ratings = {}

    matches = ratings_pattern.findall(evaluation)
    for criterion, rating in matches:
        ratings[criterion] = int(rating)

    return ratings

ratings_list = []
for result in results:
    ratings = extract_winners(result['Evaluation'])
    ratings['Question'] = result['Question']
    ratings_list.append(ratings)

ratings_df = pd.DataFrame(ratings_list)
```
